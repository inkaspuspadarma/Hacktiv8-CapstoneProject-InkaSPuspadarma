# Data Cleaning


```python
import re
import string
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
```

```python
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('wordnet')
nltk.download('punkt_tab')
```

```python
def remove_emoji(text):
    emoji_pattern = re.compile(
        "["
        u"\U0001F600-\U0001F64F"  # emotikon wajah
        u"\U0001F300-\U0001F5FF"  # simbol & pictogram
        u"\U0001F680-\U0001F6FF"  # simbol transportasi
        u"\U0001F1E0-\U0001F1FF"  # bendera negara
        u"\U00002700-\U000027BF"
        u"\U000024C2-\U0001F251"
        "]+", flags=re.UNICODE)
    return emoji_pattern.sub(r'', text)

# === Fungsi utama cleaning ===
def clean_english_text(text):
    if pd.isnull(text):
        return ""

    text = remove_emoji(text)  # 🧽 Hapus emoji dulu
    text = text.lower()
    text = re.sub(r'http\S+|www.\S+', '', text)
    text = re.sub(r'@\w+|#\w+', '', text)
    text = re.sub(r'\d+', '', text)
    text = text.translate(str.maketrans('', '', string.punctuation))
    text = re.sub(r'\s+', ' ', text).strip()

    tokens = nltk.word_tokenize(text)
    stop_words = set(stopwords.words('english'))
    tokens = [word for word in tokens if word not in stop_words]

    lemmatizer = WordNetLemmatizer()
    tokens = [lemmatizer.lemmatize(word) for word in tokens]

    return ' '.join(tokens)

# === Terapkan ke kolom content ===
df_dataused['cleaned_content'] = df_dataused['content'].apply(clean_english_text)
```

# Sentiment Classification

```python
import pandas as pd

# Asumsi df_clean sudah ada dan kolomnya bernama "content"
review_list2 = df_clean["content"].tolist()

# Simpan hasil akhir
final_results = []

# Loop per 10 review
batch_size = 5
for i in range(0, len(review_list2), batch_size):
    batch = review_list2[i:i + batch_size]

    # Buat text prompt untuk batch ini
    text_review = "\n".join([f"{r}" for r in batch])

    prompt_sentimen = f"""
You are a sentiment classification assistant.

Label the sentiment of each review below using:
- 0 = Negative (complaints, frustration, bad experience)
- 1 = Neutral (mixed opinion, okay, unclear tone)
- 2 = Positive (satisfaction, praise, good experience)

Follow the examples first, then continue labeling the provided reviews.

Examples:
Review: This app is terrible and always crashes → 0
Review: Not bad, just average. Nothing special. → 1
Review: I love this app, it's super useful! → 2

Now label the following reviews (just return the numbers in order, one per line):

{text_review}
"""

    # Kirim ke model Granite (ganti dengan method kamu sendiri)
    response = output.invoke(prompt_sentimen)

    # Pecah respons jadi list angka
    label_batch = response.strip().split("\n")

    # Gabungkan review dan label ke hasil akhir
    for review_text, label in zip(batch, label_batch):
        final_results.append({
            "review": review_text,
            "label": int(label.strip())
        })

# Simpan ke DataFrame
df_labeled2 = pd.DataFrame(final_results)

# Tampilkan hasil
print(df_labeled2)
```

# Review Category Classification

```python
import pandas as pd

# Ambil isi review
review_list2 = df_clean["content"].tolist()

# Simpan hasil akhir
final_results = []

# Parameter model (opsional, disesuaikan Granite)
parameters = {
    "top_k": 1,
    "top_p": 0.8,
    "max_tokens": 30,
    "min_tokens": 1,
    "random_seed": 42,
    "repetition_penalty": 1.5,
    "temperature": 0.3,
    "stopping_criteria": "length",
    "stopping_sequence": None
}

# Loop per 5 review
batch_size = 5
for i in range(0, len(review_list2), batch_size):
    batch = review_list2[i:i + batch_size]
    text_review = "\n".join([f"{r}" for r in batch])

    # Prompt Kategori Keluhan
    prompt_kategori = f"""
You are a customer complaint classifier.

Classify each review into one of the following categories based on the main issue:

1 = Login/Verification
2 = Payment/Top-up
3 = Crash/Bug/Error
4 = Performance/Slow
5 = UI/UX/Navigability
6 = Package/Data/Promo
7 = Other

Examples:
Review: I can't log in with my number anymore → 1
Review: Payment keeps failing for no reason → 2
Review: App crashes every time I open it → 3
Review: It's too slow and keeps loading → 4
Review: I don't understand the new layout → 5
Review: My package disappeared after payment → 6
Review: I'm frustrated with the app → 7

Now classify the following reviews (return the number only, one per line):

{text_review}
"""

    # Kirim ke model
    response = output.invoke(prompt_kategori, parameters=parameters)

    # Pecah hasil jadi list angka
    label_kategori = response.strip().split("\n")

    # Gabungkan review dan label kategori ke hasil akhir
    for review_text, label in zip(batch, label_kategori):
        final_results.append({
            "review": review_text,
            "kategori_keluhan": int(label.strip())
        })

# Simpan ke DataFrame
df_kategori = pd.DataFrame(final_results)
print(df_kategori)
```

```python
import pandas as pd

# Load data hasil klasifikasi kategori keluhan
df = pd.read_csv("complaint_category.csv")

# Mapping label ke nama kategori
kategori_map = {
    1: "Login/Verification",
    2: "Payment/Top-up",
    3: "Crash/Bug/Error",
    4: "Performance/Slow",
    5: "UI/UX/Navigability",
    6: "Package/Data/Promo",
    7: "Other"
}

summary_result = []

# Loop setiap kategori
for kategori_kode, kategori_nama in kategori_map.items():
    # Ambil review kategori ini
    reviews = df[df["kategori_keluhan"] == kategori_kode]["review"].dropna().tolist()
    if not reviews:
        continue  # skip kalau gak ada review

    # Gabungkan maksimal 50 review
    text_review = "\n".join(reviews[:50])

    # Buat prompt
    prompt_kategori_summary = f"""
You are a summarization assistant.

Below are user reviews that fall into the complaint category: "{kategori_nama}".

Summarize the 3 to 5 most common complaints found in these reviews. Be specific, group similar issues, and write clear, actionable summaries.

Reviews:
{text_review}
"""

    # Kirim ke model Granite
    result = output.invoke(prompt_kategori_summary)

    # Simpan ke list
    summary_result.append({
        "kategori": kategori_nama,
        "summary": result.strip()
    })

# Buat DataFrame & simpan ke CSV
df_summary_kategori = pd.DataFrame(summary_result)
display(df_summary_kategori)
```

```python
import pandas as pd

# Ambil 5 review dengan thumbs up terbanyak
top5_reviews = df_clean.sort_values(by='thumbsUpCount', ascending=False).dropna(subset=['content']).head(5).copy()

# Bersihkan teks review
def clean_review(text):
    text = text.strip()
    if not text.endswith('.'):
        text += '.'
    return text[0].upper() + text[1:]

top5_reviews['cleaned'] = top5_reviews['content'].apply(clean_review)

# Format untuk prompt
top5_text = "\n".join([f"{i+1}. {row}" for i, row in enumerate(top5_reviews['cleaned'])])

# Prompt untuk IBM Granite
prompt = f"""
You are a senior customer experience analyst.

You are given 5 unstructured user reviews that received the most thumbs-up. These reviews may be informal and lack punctuation.

Your task:
1. Interpret and summarize the top 3–5 main complaints raised.
2. Explain why these reviews resonated with others.
3. Recommend clear, actionable improvements for the product team.

Respond ONLY with the insights. Do NOT repeat the reviews in your response.

Format your output like this (keep it concise and structured):

🔍 Agreed Complaints:
• Complaint 1
• Complaint 2
• Complaint 3

📊 Why Users Agree:
• Reason 1
• Reason 2

✅ Recommended Actions:
• Action 1
• Action 2

[BEGIN USER REVIEWS]
{top5_text}
[END USER REVIEWS]
"""

# Kirim ke Granite (ganti dengan output.invoke jika di luar)
response = output.invoke(prompt)

# Parsing ke DataFrame
def parse_granite_output(text):
    sections = {"Agreed Complaints": [], "Why Users Agree": [], "Recommended Actions": []}
    current_section = None

    for line in text.splitlines():
        if "🔍" in line:
            current_section = "Agreed Complaints"
        elif "📊" in line:
            current_section = "Why Users Agree"
        elif "✅" in line:
            current_section = "Recommended Actions"
        elif line.startswith("•") and current_section:
            sections[current_section].append(line.replace("•", "").strip())

    # Buat tabel
    max_len = max(len(sections["Agreed Complaints"]), len(sections["Why Users Agree"]), len(sections["Recommended Actions"]))
    df_result = pd.DataFrame({
        "Agreed_Complaints": sections["Agreed Complaints"] + [""] * (max_len - len(sections["Agreed Complaints"])),
        "Why_Users_Agree": sections["Why Users Agree"] + [""] * (max_len - len(sections["Why Users Agree"])),
        "Recommended_Actions": sections["Recommended Actions"] + [""] * (max_len - len(sections["Recommended Actions"]))
    })
    return df_result

# Tampilkan hasil
df_summary_top5 = parse_granite_output(response)
display(df_summary_top5)
```

# WordCloud Implementation

```python
# --- Install libraries ---
!pip install wordcloud matplotlib contractions --quiet

# --- Import libraries ---
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud
import re
import string
import contractions
from nltk.corpus import stopwords
import nltk
nltk.download('stopwords')

# --- Load file CSV ---
df = pd.read_csv("complaint_category.csv")

# --- Validasi kolom ---
if "review" not in df.columns:
    raise ValueError("❌ Kolom 'review' tidak ditemukan dalam file.")

# --- Cleaning function ---
def clean_text(text):
    text = contractions.fix(text)
    text = text.lower()
    text = re.sub(r'[^\x00-\x7F]+', ' ', text)
    text = re.sub(r'http\S+|www.\S+', '', text)
    text = re.sub(r'\d+', '', text)
    text = text.translate(str.maketrans('', '', string.punctuation))
    text = re.sub(r'\s+', ' ', text).strip()

    stop_words = set(stopwords.words('english'))
    custom_stopwords = {"even", "make", "need", "please", "use"}  # ✅ Stopwords tambahan
    stop_words.update(custom_stopwords)

    words = [word for word in text.split() if word not in stop_words]
    return ' '.join(words)

# --- Bersihkan kolom review ---
df['cleaned_review'] = df['review'].astype(str).apply(clean_text)

# --- Gabung semua teks jadi 1 string ---
text = " ".join(df['cleaned_review'].dropna().tolist())

# --- Generate WordCloud ---
wordcloud = WordCloud(width=1000, height=500, background_color='white').generate(text)

# --- Tampilkan WordCloud ---
plt.figure(figsize=(15, 7))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title("☁️ WordCloud Komplain Pengguna (Cleaned)", fontsize=20)
plt.show()

# --- Simpan WordCloud ke file PNG ---
output_image = "wordcloud_mytelkomsel.png"
wordcloud.to_file(output_image)
print(f"✅ WordCloud disimpan sebagai {output_image}")
```
